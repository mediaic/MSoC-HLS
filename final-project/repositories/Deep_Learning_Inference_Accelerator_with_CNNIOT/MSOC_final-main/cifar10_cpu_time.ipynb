```python
import numpy as np
import time

conv1_w=np.load('cifar10_weights_v2/conv1.weight.npy')
conv1_b=np.load('cifar10_weights_v2/conv1.bias.npy')
conv2_w=np.load('cifar10_weights_v2/conv2.weight.npy')
conv2_b=np.load('cifar10_weights_v2/conv2.bias.npy')
conv3_w=np.load('cifar10_weights_v2/conv3.weight.npy')
conv3_b=np.load('cifar10_weights_v2/conv3.bias.npy')
conv4_w=np.load('cifar10_weights_v2/conv4.weight.npy')
conv4_b=np.load('cifar10_weights_v2/conv4.bias.npy')
f1_w=np.load('cifar10_weights_v2/fc1.weight.npy')
f1_b=np.load('cifar10_weights_v2/fc1.bias.npy')
f2_w=np.load('cifar10_weights_v2/fc2.weight.npy')
f2_b=np.load('cifar10_weights_v2/fc2.bias.npy')
f3_w=np.load('cifar10_weights_v2/fc3.weight.npy')
f3_b=np.load('cifar10_weights_v2/fc3.bias.npy')

data = np.load('imgs.npy')
print(data.shape)

data2 = data[0]
t1 = time.time()
#CONV1 (6,3,3)
rs = np.zeros((6, 30, 30))

for i in range(0, 6):
    for i1 in range(0, 30):
        for i2 in range(0, 30):
            tmp = data2[0, i1:i1 + 3, i2:i2 + 3]
            rs[i][i1][i2] = np.sum(np.multiply(tmp, conv1_w[i])) + conv1_b[i]

#CONV2 (10,3,3)
rs2 = np.zeros((10, 28, 28))
for i in range(0, 10):
    for i1 in range(0, 28):
        for i2 in range(0, 28):
            tmp = rs[0, i1:i1 + 3, i2:i2 + 3]
            rs2[i][i1][i2] = np.sum(np.multiply(tmp, conv2_w[i])) + conv2_b[i]
            
#POOLING
rs3 = np.zeros(( 10, 14, 14))
for i in range(0, 10):
    i1=0
    while i1<28:
        i2=0
        while i2<28:
            tmp = rs2[i, i1:i1 + 2, i2:i2 + 2]
            tmp=np.max(tmp)
#RELU
            if tmp>0:
                rs3[i][int(i1/2)][int(i2/2)] =tmp
            i2+=2
        i1+=2
        
#CONV3 (16,3,3)
rs4 = np.zeros((16, 12, 12))
for i in range(0, 16):
    for i1 in range(0, 12):
        for i2 in range(0, 12):
            tmp = rs3[0, i1:i1 + 3, i2:i2 + 3]
            rs4[i][i1][i2] = np.sum(np.multiply(tmp, conv3_w[i])) + conv3_b[i]
            
#CONV4 (20,3,3)
rs5 = np.zeros((20, 10, 10))
for i in range(0, 20):
    for i1 in range(0, 10):
        for i2 in range(0, 10):
            tmp = rs4[0, i1:i1 + 3, i2:i2 + 3]
            rs5[i][i1][i2] = np.sum(np.multiply(tmp, conv4_w[i])) + conv4_b[i]
            
#POOLING
rs6 = np.zeros(( 20, 5, 5))
for i in range(0, 20):
    i1=0
    while i1<10:
        i2=0
        while i2<10:
            tmp = rs5[i, i1:i1 + 2, i2:i2 + 2]
            tmp=np.max(tmp)
#RELU
            if tmp>0:
                rs6[i][int(i1/2)][int(i2/2)] =tmp
            i2+=2
        i1+=2
        
        
rs = np.reshape(rs6, ( 500))
print(rs.shape)

rs2 = np.zeros(120)
tmp=rs
tmp=np.add(np.dot(f1_w, tmp), f1_b)
for i in range(0,120):
#RELU
    if tmp[i]>0:
        rs2[i]=tmp[i]
        
rs3 = np.zeros(84)
tmp=rs2
tmp=np.add(np.dot(f2_w, tmp), f2_b)
for i in range(0,84):
#RELU
    if tmp[i]>0:
        rs3[i]=tmp[i]
        
rs4 = np.zeros(10)
tmp=rs3
rs4=np.add(np.dot(f3_w, tmp), f3_b)
#for i in range(0,10):
#RELU
    #if tmp[i]>0:
        #rs3[i]=tmp[i]
        
t2 = time.time()
print('Inference Time',t2-t1)
#pred = np.argmax(rs4)
#print(pred)
```
